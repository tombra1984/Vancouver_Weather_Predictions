```python

# Calculate total daily precipitation for each month
monthly_total_precipitation = v2014_selected_filled.groupby('Month')['Total Rain (mm)'].sum()

# Create a bar plot to show the total precipitation for each month
plt.figure(figsize=(10, 6))
monthly_total_precipitation.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Total Precipitation (mm)')
plt.title('Total Precipitation for Each Month')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

# Identify the month with the highest total precipitation
max_precipitation_month = monthly_total_precipitation.idxmax()
print(f"The month with the highest total precipitation is {max_precipitation_month}.")
```

```python
# Calculate total daily precipitation for each month
Daily_total_precipitation = v2014_selected_filled.groupby('Day')['Total Rain (mm)'].sum()

# Create a bar plot to show the total precipitation for each month
plt.figure(figsize=(10, 6))
Daily_total_precipitation.plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Total Rain (mm)')
plt.title('Total Rainfal according to the Day')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

# Identify the month with the highest total precipitation
max_precipitation_Day = Daily_total_precipitation.idxmax()
print(f"The month with the highest total precipitation is {max_precipitation_Day}.")
```

```python
# Calculate total daily precipitation for each month
Daily_total_Heat_Days = v2014_selected_filled.groupby('Month')['Heat Deg Days (°C)'].sum()

# Create a bar plot to show the total precipitation for each month
plt.figure(figsize=(10, 6))
Daily_total_precipitation.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Heat Deg Days (°C)')
plt.title('Heat Deg Days (°C) according to the Month')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

# Identify the month with the highest total precipitation
max_Heat_Month = Daily_total_Heat_Days.idxmax()
print(f"The month with the highest total heat days is {max_Heat_Month}.")
```
